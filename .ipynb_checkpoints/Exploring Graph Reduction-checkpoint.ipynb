```python
# Parameters
D = 10  # Travel time for required arcs
S = 5   # Normal service time
Z = 12  # Zigzag service time

# Optional arc travel times (dashed lines)
optional_arc_times = {
    #(0, 4): 10,
    (4, 1): 8,
    (2, 1): 5,
    #(4,0): 12
}

# Nodes and depot
nodes = [0, 1, 2, 3, 4]
depot = 0

# Graph definition
graph = {
    "nodes": nodes,
    "depot": depot,
    "edges": {}
}

# Add required arcs (solid lines)
for i in range(len(nodes)):
    next_node = (i + 1) % len(nodes)
    if (i, next_node) in [(0, 4), (4, 0)]:
        continue
    graph["edges"][(i, next_node)] = {
        "travel_time": D,
        "service_time_normal": S,
        "service_time_zigzag": Z,
        "required": True
    }
    graph["edges"][(next_node, i)] = {
        "travel_time": D,
        "service_time_normal": S,
        "service_time_zigzag": Z,
        "required": True
    }

# Add non-required nodes
extra_nodes = [5, 6]
graph["nodes"].extend(extra_nodes)

# Add non-required arcs (increasing complexity, not shortest paths)
graph["edges"][(0, 5)] = {
    "travel_time": 5,
    "required": False
}
graph["edges"][(5, 4)] = {
    "travel_time": 5,
    "required": False
}
graph["edges"][(6, 0)] = {
    "travel_time": 6,
    "required": False
}
graph["edges"][(4, 6)] = {
    "travel_time": 6,
    "required": False
}

    
# Add optional arcs (dashed lines)
for (u, v), travel_time in optional_arc_times.items():
    graph["edges"][(u, v)] = {
        "travel_time": travel_time,
        "required": False
    }
```

## Notes

### Original Algorithm Summary

1. **Define required nodes**  
   - Nodes that are incident to at least one required arc.

2. **Induce subgraph**  
   - Create a new graph using only the required nodes.

3. **Create a complete graph**  
   - Add arcs between every pair of required nodes.
   - Arc cost = shortest path length in the original graph.

4. **Prune indirect required paths**  
   - Remove arcs where the shortest path includes other required nodes (besides the endpoints).

5. **Assumptions**  
   - **Triangle inequality** in the edges, may not hold in reality based on speed limit varying
   - This process reduces both nodes and arcs significantly (e.g., 407 → 171 nodes, 1578 → 994 arcs).

```python
# Run
graph_temp = graph

# 1) Identify required nodes
required_nodes = find_required_nodes(graph_temp)
required_set = set(required_nodes)

# 2) Build original NetworkX graph
g_orig = build_nx_graph(graph_temp)

# 3) Compute shortest paths & lengths
sp_paths, sp_lengths = compute_all_pairs_sp(g_orig)

# 4) Start with the original required edges between the required nodes (only the required start)
preserved_edges = {
    (u, v): data
    for (u, v), data in graph_temp["edges"].items()
    if u in required_set and v in required_set and data.get("required", False)
}

# 5) Build full shortest-path candidate edges
complete_edges = build_complete_edges(required_nodes, sp_lengths)

# 6) Prune down to necessary shortcuts
shortcut_edges = prune_shortcut_edges(complete_edges, sp_paths, required_set)

# 7) Label and merge shortcuts
for (u, v), data in shortcut_edges.items():
    if (u, v) not in preserved_edges:
        data["required"] = False  # distinguish shortcuts
        preserved_edges[(u, v)] = data

# 8) Final reduced graph
reduced_graph = {
    "nodes": required_nodes,
    "depot": graph_temp["depot"],
    "edges": preserved_edges
}

# 9) Visualize or continue
visualize_graph(graph_temp, title="Original Graph", pos=custom_pos)
visualize_graph(reduced_graph, title="Reduced Graph", pos=custom_pos)
```
